## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python


cloudiness = []
country = []
date = []
humidity = []
temp = []
lat = []
lng = []
wind = []
n_cities = []
rec_counter = 0
set_counter = 0
```

```python

base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + "ca7312ffe75d380e0dfe3a8412a35f02" 
print('Beginning Data Retrieval')
print('------------------------')
for city in cities:
    query_url = base_url + "&q=" + city
    response = requests.get(query_url).json()
    if rec_counter < 50:
        rec_counter = rec_counter + 1
    else:
        set_counter=set_counter + 1
        rec_counter = 0

    #print('Processing record {} of set {} | {}'.format(rec_counter, set_counter, city))
    print(f"Processing record {rec_counter} of set {set_counter} | {city}")
    print(base_url)
    try:
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        temp.append(response['main']['temp_max'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        wind.append(response['wind']['speed'])
        n_cities.append(city)
    except:
        print("City not found!")
        pass
       
print('Data Retrieval Complete')
print('-------------------------')
```

```python
sh=mydf.loc[mydf["Lat"] < 0.0]
sh.reset_index(inplace=True)
sh.head()
```

```python
def linearreg(x, y, x_label, y_label):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    print("The r-squared is: %f" % r_value**2)
    %matplotlib inline
    plt.scatter(x,y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    
    m, b = slope,intercept
    plt.plot(x, intercept + slope*x, label = 'y = {:.2f}*x + {:.2f}'.format(slope,intercept))
   
    plt.legend(loc = 'lower right')
    plt.show()
```
